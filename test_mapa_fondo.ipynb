```python
import geopandas
import contextily as cx
```

```python
# Se carga el mapa de Magdalena
magdalena = "data/Magdalena/Magdalena_municipios.shp"

# Se lee el archvio
mapa_magdalena = geopandas.read_file(magdalena)

mapa_magdalena = mapa_magdalena.set_geometry("geometry")
mapa_magdalena = mapa_magdalena.to_crs("EPSG:4326")

mapa_magdalena.plot(
    # color='white',
    figsize=(10, 10),
    alpha=0.5,
    edgecolor='k'
)
```

```python
mapa_magdalena.crs
```

```python
df_wm = mapa_magdalena.to_crs(epsg=3857)
ax = df_wm.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax)
```

```python
ax = mapa_magdalena.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, crs=mapa_magdalena.crs)
```
